```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib  inline
```

# Short-term memory in RNN

```python
N = 100
t = 5000
alpha = 0.8  #eig of W matrix
noise = 0.1  #noise strength
```

```python
A = 1
s = A*np.random.randn(t)  #stimuli
v = np.random.rand(N)
v = v/np.linalg.norm(v)  #input weight
X = np.zeros((N,t))  #neurons
z = np.random.randn(N,t)  #noise term  ###can be further filtered here
eps = 0.2
z = z/np.std(z)*eps
```

```python
temp = np.random.randn(N,N)  #recurrent connection
u,S,vh = np.linalg.svd(temp)
S = S/S[0]*alpha**2   #controling the eigenvalue
W = u @ np.diag(S) @ vh
```

```python
for tt in range(1,t):
    X[:,tt] = W @ X[:,tt-1] + v * s[tt] + z[:,tt]   #linear RNN
```

```python
plt.imshow(X[:,0:100])
```

```python
k = 1
C = np.cov(X)
pk = np.concatenate((s[k:],s[:k]),axis=0) @ X.T
uk = np.linalg.pinv(C) @ pk
y = uk @ X
y.shape
```

```python
y = y/np.std(y)
plt.plot(y[:100],'r', alpha=0.5)
plt.hold(True)
plt.plot(s[:100], '-.')
```

```python
MSE = (y-s)**2
plt.plot(MSE)
print(np.mean(MSE))
```
