## Nonlinear process

### Langevin dynamics
$\begin{aligned} \dot { \theta } _ { 1 } & = \left( \frac { \pi } { 2 } \cdot c - c \cdot \theta _ { 1 } \right) + b \dot { \omega } _ { 1 } \\ \dot { \theta } _ { 2 } & = \left( \frac { \pi } { 10 } \cdot c - c \cdot \theta _ { 2 } \right) + b \dot { \omega } _ { 2 } \end{aligned}$

### Project to 3D
$\begin{aligned} x _ { 1 } ( t ) & = \cos \left( \theta _ { 2 } \right) \sin \left( \theta _ { 1 } \right) \\ x _ { 2 } ( t ) & = \sin \left( \theta _ { 2 } \right) \sin \left( \theta _ { 1 } \right) \\ x _ { 3 } ( t ) & = \cos \left( \theta _ { 1 } \right) \end{aligned}$

```python
TT = 5
dt = 0.01
time = np.arange(0,TT,dt)
ths = np.zeros((2,len(time)))
ths[:,0] = np.random.randn(2)
xs = np.zeros((3,len(time)))
xs[:,0] = np.random.randn(3)

c = 0.024
b = 0.005

for tt in range(0,len(time)-1):
    ths[:,tt+1] = ths[:,tt] + dt*(np.pi/np.array([2,10])*c - c*ths[:,tt]) + b*np.random.randn(2)
    xs[:,tt] = np.array([np.cos(ths[1,tt])*np.sin(ths[0,tt]), np.sin(ths[1,tt])*np.sin(ths[0,tt]), np.cos(ths[0,tt])])

xs = xs[:,:-1]
```

```python
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
ax.plot3D(xs[0,:], xs[1,:], xs[2,:])
```

```python
def K(ti,tj,sig):
    """
    Kernel for diffusion mapping
    """
    dd = np.linalg.norm(ti-tj)**2
    return np.exp(dd/-sig)
```

```python
bins = np.linspace(min(xs[0,:]),max(xs[0,:]),10)
dig = np.digitize(xs[0,:], bins) - 1 # -1 because digitize returns the index of where it is larger than the value in bins
newdata = bins[dig]

plt.plot(newdata)
plt.plot(xs[0,:])
```

```python
sig = .1
Ks = np.zeros((len(time),len(time)))
for ti in range(len(time)-1):
    for tj in range(len(time)-1):
        Ks[ti,tj] = K(xs[:,ti],xs[:,tj],sig)
```

```python
Ks = Ks+10**-5
# for kk in range(Ks.shape[0]):
#     Ks[:,kk] = Ks[:,kk]/sum(Ks[:,kk])

Ks = np.diag(np.sum(Ks,axis=1)**-0.5) @ Ks @ np.diag(np.sum(Ks,axis=1)**-0.5)
plt.imshow(Ks)
```

```python
uu,ss,vv = np.linalg.svd(Ks,full_matrices=False)
```

```python
plt.plot(ss,'-o')
```
