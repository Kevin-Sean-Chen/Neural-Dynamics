## Two-population SSN network

### $\tau_i \frac{dV_i}{dt} = -V_i + V_{rest} + h_i(t) + \eta_i(t) + \sum_{E} W_{ij}r(V_j) - \sum_{I} W_{ij}r(V_j)$

### $r(V_j) = k[V_j - V_0]_+$

```python
#parameters
tauE = 0.02 #10ms
tauI = 0.01
Vr = -70  #resting potential
V0 = -70  #rectification threshold
k = 0.3  #gain
n = 2.  #exponent
WEE = 1.25
WIE = 1.2
WEI = 0.65
WII = 0.5
sigE = 0.2
sigI = 0.1
taun = 0.05  #noise correlation
```

```python
#temporal
T = 12  #time
dt = 0.001 #1ms based
time = np.arange(0,T,dt)
### stimuli and noise
ht = np.zeros(time.shape)
ht[:int(4/dt)] = 0
ht[int(4/dt):int(8/dt)] = 2
ht[int(8/dt):] = 20
noiseE = np.arange(0,T,dt)
noiseI = np.arange(0,T,dt)
for t in range(0,len(time)-1):  #O-U prcoess with correlated noise
    noiseE[t+1] = noiseE[t] + (-noiseE[t]*dt)/taun + ((2*taun*(sigE**2))**0.5)*np.random.randn()*(dt**0.5)/taun
    noiseI[t+1] = noiseI[t] + (-noiseI[t]*dt)/taun + ((2*taun*(sigI**2))**0.5)*np.random.randn()*(dt**0.5)/taun
```

```python
#check the scaling of variance
np.std(noiseI)
```

```python
##neural dynamics
VE = np.zeros(time.shape)
VI = np.zeros(time.shape)
rE = np.zeros(time.shape)
rI = np.zeros(time.shape)
VE[0] = Vr
VI[0] = Vr
for t in range(0,len(time)-1):
    rE[t] = k*max(0,(VE[t]-V0)**n)  #supralinear rectificaton
    rI[t] = k*max(0,(VI[t]-V0)**n)
    VE[t+1] = VE[t] + dt*(-VE[t] + Vr + ht[t] + WEE*rE[t] - WEI*rI[t])/tauE + noiseE[t]*dt/tauE# + (dt**0.5)*sigE*np.random.randn()/tau
    VI[t+1] = VI[t] + dt*(-VI[t] + Vr + ht[t] + WIE*rE[t] - WII*rI[t])/tauI + noiseI[t]*dt/tauI# + (dt**0.5)*sigI*np.random.randn()/tau
    
    
```

```python
plt.plot(time,VE)
plt.hold(True)
plt.plot(time,VI)
plt.ylabel('mV')
plt.xlabel('time (s)')
plt.figure()
plt.plot(time,ht)
plt.ylabel('mV')
plt.xlabel('time (s)')
```

```python
def two_SSN(h):
    VE = np.zeros(time.shape)
    VI = np.zeros(time.shape)
    rE = np.zeros(time.shape)
    rI = np.zeros(time.shape)
    VE[0] = Vr
    VI[0] = Vr
    for t in range(0,len(time)-1):
        rE[t] = k*max(0,(VE[t]-V0)**n)  #supralinear rectificaton
        rI[t] = k*max(0,(VI[t]-V0)**n)
        VE[t+1] = VE[t] + dt*(-VE[t] + Vr + h + WEE*rE[t] - WEI*rI[t])/tauE + noiseE[t]*dt/tauE# + (dt**0.5)*sigE*np.random.randn()/tau
        VI[t+1] = VI[t] + dt*(-VI[t] + Vr + h + WIE*rE[t] - WII*rI[t])/tauI + noiseI[t]*dt/tauI# + (dt**0.5)*sigI*np.random.randn()/tau
    return rE,rI,VE,VI
```

```python
hs = np.arange(0,21,1)
measure_rate = []
measure_Vmean = []
measure_Vstd = []
for hh in hs:
    re,ri,ve,vi = two_SSN(hh)
    measure_rate.append(np.array([np.mean(re),np.mean(ri)]))
    measure_Vmean.append(np.array([np.mean(ve),np.mean(vi)]))
    measure_Vstd.append(np.array([np.std(ve),np.std(vi)]))
```

```python
plt.plot(measure_rate,'-o')
plt.xlabel('input h (mV)')
plt.ylabel('spike rate (Hz)')
plt.figure()
plt.plot(measure_Vmean,'-o')
plt.xlabel('input h (mV)')
plt.ylabel('Vmean (mV)')
plt.figure()
plt.plot(measure_Vstd,'-o')
plt.xlabel('input h (mV)')
plt.ylabel('Vstd (mV)')
```

## SSN with ring architecture

### $h_i(\theta_{stim}) = b+cA_{max}exp(\frac{cos(\theta_i-\theta_{stim})-1}{l^2_{stim}})$

```python
#parameters
NE = 50
NI = 50
ls = 45
ln = 60
lst = 60
b = 2
Am = 20
thst = 0

po = np.linspace(-np.pi,np.pi,NE)  #theta_i
```

```python
#connectivity
W = np.zeros((NE,NE))
for nn in range(0,NE):
    W[:,nn] = np.exp((np.cos(po[nn]-po)-1)/ls**2)
```

```python
Wee = W*WEE/NE
Wie = W*WIE/NI
Wei = W*WEI/NE
Wii = W*WII/NI
```

```python
WW = np.block([[Wee,-Wei],[Wie,Wii]])
plt.imshow(WW)
plt.colorbar()
```

```python
sigE = 1
sigI = 0.5
noisee = np.zeros((NE,len(np.arange(0,T,dt))))
noisei = np.zeros((NI,len(np.arange(0,T,dt))))
for t in range(0,len(time)-1):  #O-U prcoess with correlated noise
    noisee[:,t+1] = noisee[:,t] + (-noisee[:,t]*dt)/taun + ((2*taun*(sigE**2))**0.5)*np.random.randn(NE)*(dt**0.5)/taun
    noisei[:,t+1] = noisei[:,t] + (-noisei[:,t]*dt)/taun + ((2*taun*(sigI**2))**0.5)*np.random.randn(NI)*(dt**0.5)/taun
```

```python
VE = np.zeros((NE,time.shape[0]))
VI = np.zeros((NI,time.shape[0]))
rE = np.zeros((NE,time.shape[0]))
rI = np.zeros((NI,time.shape[0]))
VE[0] = Vr
VI[0] = Vr
for t in range(0,len(time)-1):
    rE[:,t] = k*np.max(np.hstack((np.zeros((NE,1)),np.expand_dims((VE[:,t]-V0)**n,axis=1))),axis=1)
    rI[:,t] = k*np.max(np.hstack((np.zeros((NI,1)),np.expand_dims((VI[:,t]-V0)**n,axis=1))),axis=1)
    VE[:,t+1] = VE[:,t] + dt*(-VE[:,t] + Vr + 0 + Wee @ rE[:,t] - Wei @ rI[:,t])/tauE + noisee[:,t]*dt/tauE
    VI[:,t+1] = VI[:,t] + dt*(-VI[:,t] + Vr + 0 + Wie @ rE[:,t] - Wii @ rI[:,t])/tauI + noisei[:,t]*dt/tauI
```
